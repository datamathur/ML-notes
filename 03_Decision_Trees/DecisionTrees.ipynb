# <center>Decision Trees</center>


```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes, load_iris
from sklearn.metrics import r2_score, accuracy_score, precision_recall_fscore_support
import pandas as pd
```

## Regression Tree

```python
data1 = load_diabetes()
x, y = data1.data, data1.target
x1, x2, y1, y2 = train_test_split(x, y, test_size=0.2, random_state=42)
rgr = DecisionTreeRegressor(random_state=108)
rgr.fit(x1, y1)
y_pred = rgr.predict(x2)
r2 = r2_score(y2, y_pred)
print(f"R2 Score: {r2}")
```

```python
plot_tree(rgr)
```

## Classification Tree

```python
data1 = load_iris()
x, y = data1.data, data1.target
x1, x2, y1, y2 = train_test_split(x, y, test_size=0.2, random_state=42)
clf = DecisionTreeClassifier(random_state=108)
clf.fit(x1, y1)
y_pred = clf.predict(x2)
acc = accuracy_score(y2, y_pred)
(pre, rec, f1, sup) = precision_recall_fscore_support(y2, y_pred, average='weighted')
print(f"Accuracy Score: {acc} \nPrecision Score: {pre} \nRecall Score: {rec} \nF1-Score: {f1}")
```

```python
plot_tree(clf)
```
